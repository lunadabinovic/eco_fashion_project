```python
def first_num(string):
    
    result = string.find(' - ') 
    result
    return int(string[:result])
    
```

```python
def last_num(string):
    spl_word = ' - '
    res = string.partition(spl_word)[2]
    return int(res)
```

```python
clothes['mean1'] = clothes['Weight in grams'].apply(first_num)

```

```python

clothes['mean2'] = clothes['Weight in grams'].apply(last_num)
clothes.head()
```

```python
clothes["actual_mean"] = (clothes.mean1 + clothes.mean2)/2
```
