```python
def lower_case(ocr_splitted):
    ocr_lower = [element.lower() for element in ocr_splitted]
    return ocr_lower
```

```python
import re
from collections import Counter

def edits1(word):
    "All edits that are one edit away from `word`."
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)
```

```python
def remove_typos(ocr_lower):
    possible_typos = set()
    for word in ocr_lower:
        possible_typos.update(edits1(word))
    return list(possible_typos)
```

```python
ocr_typos_variation = remove_typos(ocr_lower)
ocr_typos_variation
```

```python
def get_matches(ocr_typos_variation):
    all_matches = []
    for fibre in fibres_list:
        matching = [s.lower() for s in ocr_typos_variation if fibre in s.lower()]
        if matching:
            matched = [fibre, matching]
            all_matches.append(matched)
    all_matches_df = pd.DataFrame(all_matches)
    return all_matches_df
```
