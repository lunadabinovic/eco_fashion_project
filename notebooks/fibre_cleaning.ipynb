```python
import string
import re
from nltk.tokenize import word_tokenize
import nltk
```

```python
def clean_string(text):
    word_tokens = word_tokenize(text)
    punctuations = string.punctuation
    clean_string = [''.join(x for x in par if x not in punctuations) for par in word_tokens]
    lower_string = [element.lower() for element in clean_string] 
    clean_string2 = [''.join(x for x in i if x.isalpha()) for i in lower_string] 
    clean_string_final = ' '.join(clean_string2).split() 
    return clean_string_final
```

```python
import re
from collections import Counter

def edits1(word):
    "All edits that are one edit away from `word`."
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)
```

```python
def remove_typos(ocr_cleaned):
    possible_typos = set()
    for word in ocr_cleaned:
        possible_typos.update(edits1(word))
    return list(possible_typos)

```

```python
def contains(fibre_list, ocr_typos_variation):
    options = []
    for ocr_word in ocr_typos_variation: #bigram and trigram
        if ocr_word in fibre_list:
            options.append(ocr_word)
    return options
        
```

```python
from sklearn.feature_extraction.text import CountVectorizer
def n_gram_from_unigram(ocr_typos_variation):
    ocr_join = " ".join(ocr_typos_variation)
    vect = CountVectorizer(ngram_range= (1,3))
    return vect.fit([ocr_join]).vocabulary_.keys()
    
```
