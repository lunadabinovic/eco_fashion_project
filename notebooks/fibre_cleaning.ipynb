```python
import string
import re
```

```python
def clean_string(text):
    txt = text.lower()
    exclude = set(string.punctuation)
    txt = ''.join(ch for ch in txt if ch not in exclude)
    txt = re.split('\n',txt)
    txt = result = ''.join([i for i in txt if not i.isdigit()])
    return txt 
```

```python
for strings in ocr_splitted:
    ocr_splitted_cleaned = clean_string(strings)
```
