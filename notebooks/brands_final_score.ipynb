# Brands Final score 

```python
# importing
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the excel file
df_final_score = pd.read_excel('../raw_data/brands_final_score.xlsx')
```

```python
df_final_score.head()
```

```python
df_final_score.drop(index=[251, 252, 253, 254], inplace =True )
```

```python
#dropping the rows with nans as a score for every section 
df_final_score.drop(index=[20, 16, 76, 102,115, 153, 250], inplace =True )
```

```python
df_final_score.columns
```

```python
# renaming the columns for the visualization for the customer 
df_final_score = df_final_score.rename(columns={'2020 Final scores - SECTION 1': 'POLICY & COMMITMENTS',
                                                '2020 FINAL PERCENTAGES -  SECTION 1': '%POLICY & COMMITMENTS',
                                                '2020 Final scores - SECTION 2': 'GOVERNANCE', 
                                                '2020 FINAL PERCENTAGES -  SECTION 2': '%GOVERNANCE', 
                                                '2020 Final scores - SECTION 3': 'TRACEABILITY',
                                               '2020 FINAL PERCENTAGES -  SECTION 3' : '%TRACEABILITY',
                                               '2020 Final scores - SECTION 4' : 'KNOW, SHOW & FIX',
                                               '2020 FINAL PERCENTAGES -  SECTION 4': '%KNOW, SHOW & FIX',
                                               '2020 Final scores - SECTION 5' : 'CONDITIONS,CONSUMPTION, CLIMATE',
                                               '2020 FINAL PERCENTAGES -  SECTION 5': '%CONDITIONS,CONSUMPTION, CLIMATE',
                                               '2020 Final scores': '2020 FINAL SCORES',
                                                '2020 FINAL PERCENTAGES': '%FINAL PERCENTAGES'})
```

```python
df_final_score.head()
```

```python
# distplot of the final 2020 score 
sns.distplot(df_final_score['%FINAL PERCENTAGES'])
```

```python
sns.histplot(df_final_score['%FINAL PERCENTAGES'])
```

```python
# only taking the Percentage information
row2 = df_final_score.iloc[0][2:11:2]
```

```python
# ploting the percentage info for the customer to see in which area the brand is 'good' and in which not
row2.plot(kind= 'bar', title = row[0])
```

```python
# function to  plot the brand that the user is inputting 
# so if he inputr 'Zara' show the graph for Zara 
def plot(df):
    index = 0 
    for row in df:
        row = df_final_score.iloc[index]
        row2= row[2:11:2]
        return row2.plot(kind= 'bar', title = row[0])
```
